```python
import numpy as np
import pandas as pd
```

```python
dgf = pd.DataFrame(np.zeros((20,1)))
dgf.index = pd.read_excel('Classifica.xlsx')['Squadra']
dgs = dgf.copy()

for n in range(26):
    df = pd.read_excel('Results22.xlsx', sheet_name = n, index_col=0)
    gg = pd.DataFrame(np.zeros((20,3)))
    gg.columns = ['Squadra', 'Gf', 'Gs']
    for i in df.index:
        gg.iloc[i, 0] = df.iloc[i, 0]
        gg.iloc[i+10, 0] = df.iloc[i, 1]
        gg.iloc[i, 1] = int(df.iloc[i, 2].split('-')[0])
        gg.iloc[i, 2] = int(df.iloc[i, 2].split('-')[1])
        gg.iloc[i+10, 1] = int(df.iloc[i, 2].split('-')[1])
        gg.iloc[i+10, 2] = int(df.iloc[i, 2].split('-')[0])
    gg.set_index('Squadra', inplace=True)
    dgf = pd.concat([dgf, gg['Gf']], axis=1)
    dgs = pd.concat([dgs, gg['Gs']], axis=1)
dgf.drop(0, axis=1, inplace=True)
dgs.drop(0, axis=1, inplace=True)
dgf.columns = [i for i in range(1, 27)]
dgs.columns = [i for i in range(1, 27)]
```

```python
with pd.ExcelWriter('Gols22.xlsx') as writer:
    for n in range(26):
        cc = pd.DataFrame(np.zeros((20, 4)))
        cc.index = dgf.index
        cc[0] = dgf.loc[:, 1:n+1].mean(axis=1)
        cc[1] = dgf.loc[:, 1:n+1].std(axis=1)
        cc[2] = dgs.loc[:, 1:n+1].mean(axis=1)
        cc[3] = dgs.loc[:, 1:n+1].std(axis=1)
        cc.fillna(0, inplace=True)
        cc.columns = ['Mgf', 'Sgf', 'Mgs', 'Sgs']
        cc.to_excel(writer, sheet_name=str(n))
```
