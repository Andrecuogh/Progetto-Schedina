```python
import numpy as np
import pandas as pd
from sklearn.naive_bayes import ComplementNB
from sklearn.multioutput import MultiOutputClassifier
from Forecasting import Forecasts
```

```python
# Importare i dataframe
dataframe = pd.read_csv('Dataframe.csv', index_col=0)

names = pd.read_excel('Previsioni.xlsx', index_col=0)
Xnot = Forecasts(26, names).drop(['Partita'], axis=1)
```

```python
# Allenare il modello statistico: naive bayes Complement
scores = []
X = dataframe.drop(['yGf', 'yGs', 'Partita'], axis=1).copy()
y = dataframe[['yGf', 'yGs']].copy()
bay = ComplementNB()       
multi_nb = MultiOutputClassifier(bay)
multi_nb.fit(X, y)
y_pred = multi_nb.predict_proba(Xnot)
scores.append(y_pred)
```

```python
# Probabilit√† di fare x gol
pd.DataFrame(scores[0][0]).set_index(names[0])
```

```python
# Probabilit√† di subire x gol
pd.DataFrame(scores[0][1]).set_index(names[0])
```
