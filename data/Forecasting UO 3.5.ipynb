```python
import numpy as np
import pandas as pd
from sklearn.naive_bayes import ComplementNB
from Forecasting import Forecasts
```

```python
# Importare i dataset
dataframe = pd.read_csv('Dataframe.csv', index_col=0)
names = pd.read_excel('Previsioni.xlsx', index_col=0)
Xnot = Forecasts(25, names).drop(['Partita'], axis=1)

# Creare la variabile 'Classe' per UO 3.5
dataframe['Classe'] = 0
for i in dataframe.index:
    if dataframe.loc[i, 'yGf'] + dataframe.loc[i, 'yGs'] > 3:
        dataframe.loc[i, 'Classe'] = 'O'
    else:
        dataframe.loc[i, 'Classe'] = 'U'
dataframe.drop(['yGf', 'yGs'], axis=1, inplace=True)

dataframe
```

```python
# Allenare il modello: naive bayes complement
scores = []
X = dataframe.drop(['Classe', 'Partita'], axis=1).copy()
y = dataframe['Classe'].copy()
reg = ComplementNB(alpha=0.1)
reg.fit(X, y)
y_pred = reg.predict_proba(Xnot)
scores.append(y_pred)
```

```python
# probabilità di Over Under 3.5
pd.DataFrame(scores[0], columns = reg.classes_).set_index(names[0])
```
