```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from Forecasting import Forecasts
```

```python
# Importare i dataset
dataframe = pd.read_csv('Dataframe.csv', index_col=0)
names = pd.read_excel('Previsioni.xlsx', index_col=0)
Xnot = Forecasts(26, names).drop(['Partita'], axis=1)

# Creare la variabile Classe per GG o NG
dataframe['Classe'] = 0
for i in dataframe.index:
    if dataframe.loc[i, 'yGf'] != 0 and dataframe.loc[i, 'yGs'] != 0:
        dataframe.loc[i, 'Classe'] = 'GG'
    else:
        dataframe.loc[i, 'Classe'] = 'NG'
dataframe.drop(['yGf', 'yGs'], axis=1, inplace=True)

dataframe.head()
```

```python
# Allenare il modello: Regressione Logistica
scores = []
X = dataframe.drop(['Classe', 'Partita'], axis=1).copy()
y = dataframe['Classe'].copy()
reg = LogisticRegression(max_iter = 1000)
reg.fit(X, y)
y_pred = reg.predict_proba(Xnot)
scores.append(y_pred)
```

```python
# Probabilità di GOL e NO GOL
pd.DataFrame(scores[0], columns = reg.classes_).set_index(names[0])
```
