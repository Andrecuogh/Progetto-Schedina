```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LogisticRegression
from Forecasting import Forecasts
```

```python
# Importare i dataframe
dataframe = pd.read_csv('Dataframe.csv', index_col=0)
names = pd.read_excel('Previsioni.xlsx', index_col=0)
Xnot = Forecasts(25, names).drop(['Partita'], axis=1)

# Trasformare i gol fatti-subiti in risultato 'V', 'N', 'P'
dataframe['Classe'] = 0
for i in dataframe.index:
    if dataframe.loc[i, 'yGf'] > dataframe.loc[i, 'yGs']:
        dataframe.loc[i, 'Classe'] = 'V'
    elif dataframe.loc[i, 'yGf'] == dataframe.loc[i, 'yGs']:
        dataframe.loc[i, 'Classe'] = 'N'
    else:
        dataframe.loc[i, 'Classe'] = 'P'
dataframe.drop(['yGf', 'yGs'], axis=1, inplace=True)

dataframe.head()
```

```python
# Allenare il modello: Regressione Logistica Multinomiale
scores = []
X = dataframe.drop(['Classe', 'Partita'], axis=1).copy()
y = dataframe['Classe'].copy()
reg = LogisticRegression(max_iter = 1000, multi_class = 'multinomial')
reg.fit(X, y)
y_pred = reg.predict_proba(Xnot)
scores.append(y_pred)
```

```python
# Probabilità dell'esito della partita
pd.DataFrame(scores[0], columns = reg.classes_).set_index(names[0])
```
