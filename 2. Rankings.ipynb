```python
import pandas as pd
import numpy as np
```

```python
with pd.ExcelWriter('Stand22.xlsx') as writer: 
    cltot = pd.DataFrame(np.zeros((20, 8)))
    cltot.columns = ['Squadra', 'Gf', 'Gs', 'Punti', 'V', 'N', 'P', 'Diff']
    cltot['Squadra'] = ['Milan', 'Sampdoria', 'Lecce', 'Monza', 'Fiorentina', 
                        'Lazio', 'Salernitana', 'Spezia', 'Verona', 'Juventus', 
                        'Udinese', 'Atalanta', 'Inter', 'Torino', 'Cremonese',
                        'Bologna', 'Roma', 'Empoli', 'Napoli', 'Sassuolo']
    cltot.set_index('Squadra', inplace=True)

    for n in range(26):
        df = pd.read_excel('Results22.xlsx', sheet_name=n, index_col=0)
        a, b = df[2].values[0].split('-')
        classifica = pd.DataFrame(np.zeros((20,3), dtype=np.int64))
        for i in range(10):
            classifica.iloc[i, 0] = df.iloc[i,0]
            classifica.iloc[i, 1] = int(df.iloc[i, 2].split('-')[0])
            classifica.iloc[i, 2] = int(df.iloc[i, 2].split('-')[1])
            classifica.iloc[i+10, 0] = df.iloc[i, 1]
            classifica.iloc[i+10, 1] = int(df.iloc[i, 2].split('-')[1])
            classifica.iloc[i+10, 2] = int(df.iloc[i, 2].split('-')[0])
        classifica.columns = ['Squadra', 'Gf', 'Gs']
        classifica.set_index('Squadra', inplace=True)
        classifica['Punti'] = 0
        classifica['V'] = 0
        classifica['N'] = 0
        classifica['P'] = 0
        for indx, team in enumerate(classifica.index):
            if classifica.loc[team, 'Gf'] > classifica.loc[team, 'Gs']:
                classifica.iloc[indx, 2] = 3
                classifica.iloc[indx, 3] = 1
            elif classifica.loc[team, 'Gf'] == classifica.loc[team, 'Gs']:
                classifica.iloc[indx, 2] = 1
                classifica.iloc[indx, 4] = 1
            else:
                classifica.iloc[indx, 5] = 1
        classifica['Diff'] = classifica['Gf'] - classifica['Gs']
        cltot = classifica + cltot
        cltot.sort_values(by=['Punti', 'Diff', 'Gf', 'Squadra'], ascending = [False, False, False, True], inplace=True)
        cltot['Posizione'] = np.arange(1, 21)
        cltot.to_excel(writer, sheet_name = str(n))
```
